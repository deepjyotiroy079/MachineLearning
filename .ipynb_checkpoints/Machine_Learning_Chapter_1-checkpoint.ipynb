# scipy

```python
# SciPy is a collection of functions for scientific computing in Python. It provides,
# among other functionality, advanced linear algebra routines, mathematical function
# optimization, signal processing, special mathematical functions, and statistical distri‐
# butions. scikit-learn draws from SciPy’s collection of functions for implementing
# its algorithms.
```

```python
from scipy import sparse
```

```python
# Create a 2D NumPy array with a diagonal of ones, and zeros everywhere else
eye = np.eye(4)
print("NumPy array:\n{}".format(eye))
```

#### 

```python
# keep a reference of https://www.geeksforgeeks.org/sparse-matrix-representations-set-3-csr/ for CSR representation of sparse matrix
```

```python
# creating a sparse matrix in csr format
sparse_matrix = sparse.csr_matrix(eye)
print("Scipy Sparse matrix in CSR format :\n\n{}".format(sparse_matrix))
```

```python
# Coo Format (Co-ordinative Format)
# data, rows, cols
# https://scipy-lectures.org/advanced/scipy_sparse/coo_matrix.html
```

```python
data = np.ones(4)
row_indices = np.arange(4)
col_indices = np.arange(4)
eye_coo = sparse.coo_matrix((data, (row_indices, col_indices)))
print("COO representation:\n{}".format(eye_coo))
```

# First Machine Learning Program (The IRIS dataset)

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
# printing the keys in the iris dataset
```

```python
print("Key's in the Iris dataset : \n{}".format(iris_dataset.keys()))
```

```python
print(iris_dataset['DESCR'][:193]+"\n...")
```

```python
# get the target names
print("target names : \n{}".format(iris_dataset['target_names']))
```

```python
# get the feature names
print("feature names : \n{}".format(iris_dataset['feature_names']))
```

```python
# get the type of data
print("Type of Data : {}".format(type(iris_dataset['data'])))
```

```python
#  The shape of the data array is the number of samples multiplied by the number of features
print("Shape of the data : {}".format(iris_dataset['data'].shape))
```

```python
# getting first five data of the dataset
print("First five data of the dataset : \n{}".format(iris_dataset['data'][:5]))
```

# Remember that the individual items are called samples in machine learning, and their properties are called features.

```python
# getting the datatype of target
print("Target data type: {}".format(iris_dataset['target'].shape))
```

```python
print("Target data type: {}".format(iris_dataset['target']))
```

```python
# scikit-learn contains a function that shuffles the dataset and splits it for you: the
# train_test_split function. This function extracts 75% of the rows in the data as the
# training set, together with the corresponding labels for this data. The remaining 25%
# of the data, together with the remaining labels, is declared as the test set. 

# In scikit-learn, data is usually denoted with a capital X, while labels are denoted by
# a lowercase y. This is inspired by the standard formulation f(x)=y in mathematics,
# where x is the input to a function and y is the output. 
```

```python
# breaking the dataset into two parts, one for training and one for testing
from sklearn.model_selection import train_test_split
```

```python
# this randomly shuffles the dataset
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
# X_train contains 75% data and X_test contains 25
print("X_train data : {}".format(X_train.shape))
print("y_train data : {}".format(y_train.shape))
```

```python
# the testing data
print("X_test data : {}".format(X_test.shape))
print("y_test data : {}".format(y_test.shape))
```

```python
# Create a DataFrame from given numpy arrays of X_train, y_train, X_test, y_test
import mglearn
```

```python
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
```

```python
grr = pd.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8, cmap = mglearn.cm3)
```

# K Nearest Neighbor models

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# initially the number of neighbors is set to 1
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
# knn.fit() returns the knn object itself, we get a string representation of our classifier
knn.fit(X_train, y_train)
```

# Making Predictions

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
print("X_new.shape : {}".format(X_new.shape))
```

# to perform the prediction we call the predict method of the knn object

```python
prediction = knn.predict(X_new)
print("Prediction : {}".format(prediction))
print("Prediction target name : {}".format(iris_dataset['target_names'][prediction]))
```

# Evalutating the Model

```python
# This is where the test set that we created earlier comes in. This data was not used to
# build the model, but we do know what the correct species is for each iris in the test
# set.

# Therefore, we can make a prediction for each iris in the test data and compare it
# against its label (the known species). We can measure how well the model works by
# computing the accuracy, which is the fraction of flowers for which the right species
# was predicted
```

```python
y_pred = knn.predict(X_test)
print("Test set Prediction : {}".format(y_pred))
```

```python
print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("Test score of the knn object : \n\n{}".format(knn.score(X_test, y_test)))
```
